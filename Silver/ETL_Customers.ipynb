```python
from pyspark.sql import functions as F
from pyspark.sql.functions import col
```

#####Customers Detla table

```python
df_customers=spark.read.table('olist.bronze.crm_olist_customers')
```

```python
df_customers.display()
```

##### check for white sapce

```python
%sql
SELECT *
FROM olist.bronze.crm_olist_customers
WHERE customer_state != TRIM(customer_state); 
```

##### Change the data type of custmer_zip_code_prefix

```python
df_customers=df_customers.withColumn("customer_zip_code_prefix", col("customer_zip_code_prefix").cast("int"))
```
